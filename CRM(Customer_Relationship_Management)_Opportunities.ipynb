## Unzip the file

### Subtask:
Unzip the provided zip file `/content/CRM.zip` to access its contents.


**Reasoning**:
Unzip the provided zip file `/content/CRM.zip` to access its contents.



```python
import zipfile

zip_file_path = '/content/CRM.zip'
extraction_dir = '/content/'

with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extraction_dir)

import os
print(os.listdir(extraction_dir))
```

## Load the dataset

### Subtask:
Load the dataset into a pandas DataFrame.


**Reasoning**:
Import the pandas library and load the CSV files into separate DataFrames as instructed.



```python
import pandas as pd

df_sales_teams = pd.read_csv('/content/sales_teams.csv')
df_accounts = pd.read_csv('/content/accounts.csv')
df_data_dictionary = pd.read_csv('/content/data_dictionary.csv')
df_products = pd.read_csv('/content/products.csv')
df_sales_pipeline = pd.read_csv('/content/sales_pipeline.csv')

print("Sales Teams DataFrame:")
display(df_sales_teams.head())

print("\nAccounts DataFrame:")
display(df_accounts.head())

print("\nData Dictionary DataFrame:")
display(df_data_dictionary.head())

print("\nProducts DataFrame:")
display(df_products.head())

print("\nSales Pipeline DataFrame:")
display(df_sales_pipeline.head())
```

## Perform data visualization

### Subtask:
Create visualizations to represent the data and insights gained from the descriptive and exploratory analysis.


**Reasoning**:
Generate the required visualizations based on the instructions, grouping related plots for clarity and efficiency.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# 1. Bar plot for 'sector' in df_accounts
plt.figure(figsize=(10, 6))
sns.countplot(data=df_accounts, y='sector', order=df_accounts['sector'].value_counts().index, palette='viridis')
plt.title('Distribution of Accounts by Sector')
plt.xlabel('Number of Accounts')
plt.ylabel('Sector')
plt.tight_layout()
plt.show()

# 2. Bar plot for 'office_location' in df_accounts
plt.figure(figsize=(10, 8))
sns.countplot(data=df_accounts, y='office_location', order=df_accounts['office_location'].value_counts().index, palette='viridis')
plt.title('Distribution of Accounts by Office Location')
plt.xlabel('Number of Accounts')
plt.ylabel('Office Location')
plt.tight_layout()
plt.show()

# 3. Histogram of 'revenue' in df_accounts
plt.figure(figsize=(10, 6))
sns.histplot(data=df_accounts, x='revenue', bins=20, kde=True, color='skyblue')
plt.title('Distribution of Account Revenue')
plt.xlabel('Revenue (in millions of USD)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# 4. Histogram of 'employees' in df_accounts
plt.figure(figsize=(10, 6))
sns.histplot(data=df_accounts, x='employees', bins=20, kde=True, color='lightcoral')
plt.title('Distribution of Account Employees')
plt.xlabel('Number of Employees')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# 5. Bar plot for 'deal_stage' in df_sales_pipeline
plt.figure(figsize=(10, 6))
sns.countplot(data=df_sales_pipeline, y='deal_stage', order=df_sales_pipeline['deal_stage'].value_counts().index, palette='viridis')
plt.title('Distribution of Deals by Stage')
plt.xlabel('Number of Deals')
plt.ylabel('Deal Stage')
plt.tight_layout()
plt.show()

# 6. Histogram of 'close_value' in df_sales_pipeline (handling missing values)
plt.figure(figsize=(10, 6))
sns.histplot(data=df_sales_pipeline.dropna(subset=['close_value']), x='close_value', bins=20, kde=True, color='skyblue')
plt.title('Distribution of Deal Close Values')
plt.xlabel('Close Value')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# 7. Convert date columns to datetime objects
df_sales_pipeline['engage_date'] = pd.to_datetime(df_sales_pipeline['engage_date'])
df_sales_pipeline['close_date'] = pd.to_datetime(df_sales_pipeline['close_date'])

# 8. Create 'deal_duration' column
df_sales_pipeline['deal_duration'] = (df_sales_pipeline['close_date'] - df_sales_pipeline['engage_date']).dt.days

# 9. Histogram of 'deal_duration' (handling missing values)
plt.figure(figsize=(10, 6))
sns.histplot(data=df_sales_pipeline.dropna(subset=['deal_duration']), x='deal_duration', bins=20, kde=True, color='lightcoral')
plt.title('Distribution of Deal Durations')
plt.xlabel('Deal Duration (days)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# 10. Bar plot for 'regional_office' in df_sales_teams
plt.figure(figsize=(10, 6))
sns.countplot(data=df_sales_teams, y='regional_office', order=df_sales_teams['regional_office'].value_counts().index, palette='viridis')
plt.title('Distribution of Sales Teams by Regional Office')
plt.xlabel('Number of Sales Teams')
plt.ylabel('Regional Office')
plt.tight_layout()
plt.show()

# 11. Bar plot for 'manager' in df_sales_teams
plt.figure(figsize=(10, 8))
sns.countplot(data=df_sales_teams, y='manager', order=df_sales_teams['manager'].value_counts().index, palette='viridis')
plt.title('Distribution of Sales Teams by Manager')
plt.xlabel('Number of Sales Teams')
plt.ylabel('Manager')
plt.tight_layout()
plt.show()

# 12. Bar plot for 'series' in df_products
plt.figure(figsize=(10, 6))
sns.countplot(data=df_products, y='series', order=df_products['series'].value_counts().index, palette='viridis')
plt.title('Distribution of Products by Series')
plt.xlabel('Number of Products')
plt.ylabel('Series')
plt.tight_layout()
plt.show()

# 13. Histogram of 'sales_price' in df_products
plt.figure(figsize=(10, 6))
sns.histplot(data=df_products, x='sales_price', bins=5, kde=True, color='skyblue')
plt.title('Distribution of Product Sales Prices')
plt.xlabel('Sales Price')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
