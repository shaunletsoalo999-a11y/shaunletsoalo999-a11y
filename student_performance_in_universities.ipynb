## Load data

### Subtask:
Load the dataset "Cleaned_Students_Performance.csv" into a pandas DataFrame.


**Reasoning**:
Load the dataset into a pandas DataFrame and display the first few rows.



```python
import pandas as pd
df = pd.read_csv("/content/Cleaned_Students_Performance.csv")
display(df.head())
```

## Descriptive statistics

### Subtask:
Calculate and display descriptive statistics for the numerical columns in the dataset.


**Reasoning**:
Calculate and display descriptive statistics for the numerical columns in the DataFrame.



```python
numerical_cols = df.select_dtypes(include=['number'])
descriptive_stats = numerical_cols.describe()
display(descriptive_stats)
```

## Data information

### Subtask:
Display information about the dataset, including data types and non-null values.


**Reasoning**:

## Check for missing values

### Subtask:
Check for and visualize any missing values in the dataset.


**Reasoning**:
Check for missing values in each column and print a message if none are found, as the previous df.info() indicated no missing values.



```python
missing_values = df.isnull().sum()

if missing_values.sum() == 0:
    print("No missing values found in the dataset.")
else:
    print("Missing values found:\n", missing_values)
```

## Univariate analysis (numerical)

### Subtask:
Create histograms and box plots for numerical columns to understand their distributions.


**Reasoning**:
Create histograms and box plots for the numerical columns to visualize their distributions and identify outliers.



```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = df.select_dtypes(include=['number']).columns

for col in numerical_cols:
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col])
    plt.title(f'Box plot of {col}')
    plt.xlabel(col)

    plt.tight_layout()
    plt.show()
```

## Univariate analysis (categorical)

### Subtask:
Create count plots for categorical columns to visualize the frequency of each category.


**Reasoning**:
Create count plots for each categorical column to visualize the frequency distribution of their categories.



```python
categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, y=col) # Use y for horizontal bars
    plt.title(f'Count Plot of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()
```

## Bivariate analysis (numerical vs. numerical)

### Subtask:
Create scatter plots to explore relationships between numerical columns.


**Reasoning**:
Create scatter plots for all unique pairs of numerical columns to visualize their relationships.



```python
numerical_cols = df.select_dtypes(include=['number'])

for i in range(len(numerical_cols.columns)):
    for j in range(i + 1, len(numerical_cols.columns)):
        col1 = numerical_cols.columns[i]
        col2 = numerical_cols.columns[j]
        plt.figure(figsize=(10, 6))
        sns.scatterplot(data=df, x=col1, y=col2)
        plt.title(f'Scatter Plot of {col1} vs. {col2}')
        plt.xlabel(col1)
        plt.ylabel(col2)
        plt.show()
```

## Bivariate analysis (categorical vs. numerical)

### Subtask:
Create box plots or violin plots to compare numerical variables across different categories.


**Reasoning**:
Identify categorical and numerical columns, then iterate through each pair to create and display box plots.



```python
import matplotlib.pyplot as plt
import seaborn as sns

categorical_cols = df.select_dtypes(include='object').columns
numerical_cols = df.select_dtypes(include='number').columns

for cat_col in categorical_cols:
    for num_col in numerical_cols:
        plt.figure(figsize=(12, 6))
        sns.boxplot(data=df, x=cat_col, y=num_col)
        plt.title(f'Box Plot of {num_col} by {cat_col}')
        plt.xlabel(cat_col)
        plt.ylabel(num_col)
        plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
        plt.tight_layout()
        plt.show()
```

## Correlation analysis

### Subtask:
Calculate and visualize the correlation matrix for numerical columns.


**Reasoning**:
Calculate and visualize the correlation matrix for numerical columns.



```python
numerical_cols = df.select_dtypes(include=['number'])
correlation_matrix = numerical_cols.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Columns')
plt.show()
```

## Pairwise plots

### Subtask:
Generate pairwise plots (e.g., using `seaborn.pairplot`) to visualize relationships between multiple numerical variables.


**Reasoning**:
Generate a pairwise plot for the selected numerical columns using seaborn.pairplot and display the generated pairwise plot.



```python
import seaborn as sns
import matplotlib.pyplot as plt

numerical_cols = df.select_dtypes(include=['number'])

sns.pairplot(numerical_cols)
plt.suptitle('Pairwise Plots of Numerical Columns', y=1.02)
plt.show()
```
