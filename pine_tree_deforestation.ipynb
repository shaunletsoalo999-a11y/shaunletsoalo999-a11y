# Task
Generate a synthetic dataset for pine tree deforestation from January to August 2025, perform exploratory data analysis, and visualize the data.

## Generate synthetic data

### Subtask:
Create a synthetic dataset containing columns for date, region, and deforestation rate for pine trees from January to August 2025.


**Reasoning**:
The first step is to import the necessary libraries and generate the synthetic data as per the instructions.



```python
import pandas as pd
import numpy as np

dates = pd.date_range(start='2025-01-01', end='2025-08-31', freq='D')
regions = ['North', 'South', 'East', 'West']

data = {
    'Date': [],
    'Region': [],
    'Deforestation Rate': []
}

for date in dates:
    for region in regions:
        data['Date'].append(date)
        data['Region'].append(region)
        data['Deforestation Rate'].append(np.random.rand() * 0.1)

df = pd.DataFrame(data)
display(df.head())
```

## Exploratory data analysis (eda)

### Subtask:
Perform basic EDA on the dataset, including checking data types, summary statistics, and missing values.


**Reasoning**:
Display the data types, descriptive statistics, and missing values of the dataframe to perform basic EDA.



```python
display(df.info())
display(df.describe())
display(df.isnull().sum())
```

## Data visualization

### Subtask:
Create visualizations to explore the deforestation trends over time and across different regions.


**Reasoning**:
Import the necessary libraries for plotting and create the specified visualizations.



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Date', y='Deforestation Rate')
plt.title('Deforestation Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Deforestation Rate')
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Region', y='Deforestation Rate')
plt.title('Deforestation Rate Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Deforestation Rate')
plt.show()
```
