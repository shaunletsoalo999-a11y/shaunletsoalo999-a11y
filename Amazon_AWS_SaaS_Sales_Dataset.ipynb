## Load the data

### Subtask:
Load the data from "/content/SaaS-Sales.csv" into a pandas DataFrame.


**Reasoning**:
Load the data from the CSV file into a pandas DataFrame and display the first few rows to verify the load.



```python
import pandas as pd

df = pd.read_csv('/content/SaaS-Sales.csv')
display(df.head())
```

## Initial data inspection

### Subtask:
Display the first few rows, check data types, and look for missing values.


**Reasoning**:
Display the first few rows, check data types, and look for missing values in the DataFrame.



```python
display(df.head())
display(df.info())
display(df.isnull().sum())
```

## Descriptive statistics

### Subtask:
Calculate and display descriptive statistics for numerical columns.


**Reasoning**:
Calculate and display descriptive statistics for numerical columns using `.describe()` and transpose the output for better readability.



```python
display(df.describe().T)
```

## Explore categorical variables

### Subtask:
Analyze the distribution of categorical variables.


**Reasoning**:
Identify and calculate the frequency distribution of categorical columns in the DataFrame.



```python
categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    print(f"Frequency distribution for column: {col}")
    display(df[col].value_counts())
    print("\n")
```

## Data visualization

### Subtask:
Create visualizations to understand the distributions and relationships within the data. This may include histograms, box plots, scatter plots, and bar plots.


**Reasoning**:
Generate histograms and box plots for the numerical columns and bar plots for selected categorical columns.



```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = ['Sales', 'Quantity', 'Discount', 'Profit']
categorical_cols_to_plot = ['Country', 'Region', 'Segment', 'Industry']

# Histograms for numerical columns
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 2, i + 1)
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()

# Box plots for numerical columns
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(y=df[col])
    plt.title(f'Box Plot of {col}')
plt.tight_layout()
plt.show()

# Bar plots for categorical columns (showing top N categories)
plt.figure(figsize=(18, 12))
for i, col in enumerate(categorical_cols_to_plot):
    plt.subplot(2, 2, i + 1)
    top_n = df[col].value_counts().nlargest(10) # Show top 10 categories
    sns.barplot(x=top_n.index, y=top_n.values)
    plt.title(f'Frequency of {col} (Top 10)')
    plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Scatter plots for relationships between numerical variables
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.scatterplot(x=df['Sales'], y=df['Profit'])
plt.title('Sales vs Profit')

plt.subplot(1, 2, 2)
sns.scatterplot(x=df['Sales'], y=df['Quantity'])
plt.title('Sales vs Quantity')

plt.tight_layout()
plt.show()
```

## Correlation analysis

### Subtask:
Calculate and visualize the correlation matrix for numerical features.


**Reasoning**:
Calculate and visualize the correlation matrix for numerical features as instructed.



```python
numerical_cols = ['Sales', 'Quantity', 'Discount', 'Profit', 'Date Key']
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
