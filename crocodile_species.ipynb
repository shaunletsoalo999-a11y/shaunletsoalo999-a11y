```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot of Observed Length vs. Observed Weight
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Observed Length (m)', y='Observed Weight (kg)', alpha=0.6)
plt.title('Observed Length vs. Observed Weight')
plt.xlabel('Observed Length (m)')
plt.ylabel('Observed Weight (kg)')
plt.grid(True)
plt.show()

# Bar plot of Conservation Status distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Conservation Status', order=df['Conservation Status'].value_counts().index, palette='viridis')
plt.title('Distribution of Conservation Status')
plt.xlabel('Count')
plt.ylabel('Conservation Status')
plt.show()
```
